```python
def film_info(film):
    film_info = []
    csv_row = []
    
    film_info.append(film) #1 CODIGO 
    try:
        actor1 = browser.find_element_by_xpath("//h4[text()='Stars:']/following-sibling::a") #ACTOR1
        actor1_name = actor1.text
        film_info.append(actor1_name)
    except NoSuchElementException:
        actor1_name = "unknown"
        film_info.append(actor1_name) 

    try:
        actor2 = browser.find_element_by_xpath("//h4[text()='Stars:']/following-sibling::a[2]") #ACTOR2
        actor2_name = actor2.text
        film_info.append(actor2_name)
    except NoSuchElementException:
        actor2_name = "unknown"
        film_info.append(actor2_name) 

    try:
        actor3 = browser.find_element_by_xpath("//h4[text()='Stars:']/following-sibling::a[3]") #ACTOR3
        actor3_name = actor3.text
        film_info.append(actor3_name)
    except NoSuchElementException:
        actor3_name = "unknown"
        film_info.append(actor3_name) 


    if actor1_name != "unknown":

        actor1_link = actor1.get_attribute("href")

        actors_browser.get(actor1_link)
        try:
            facebook_link = actors_browser.find_element_by_link_text('Facebook')  #ACTOR1 FACEBOOK
            facebook_link.click()
            likes = actors_browser.find_element_by_xpath("//div[contains(text(), 'personas les gusta esto')]")
            head, sep, tail = likes.text.partition(' personas')
            head2, sep, tail = head.partition('A ')
            if tail:
                film_info.append(tail)
            if not tail:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown") 

        actors_browser.get(actor1_link)
        try:
            instagram_link = actors_browser.find_element_by_link_text('Instagram') #ACTOR1 INSTAGRAM
            instagram_link.click()
            seguidores = actors_browser.find_elements_by_xpath("//span[@class='g47SY ']")
            if seguidores:
                film_info.append(seguidores[1].text)
            if not seguidores:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown") 

        actors_browser.get(actor1_link)    
        try:
            twitter_link = actors_browser.find_element_by_link_text('Twitter') #ACTOR1 TWITTER
            twitter_link.click()
            time.sleep(1)
            seguidores = actors_browser.find_element_by_xpath("//div[@class='css-1dbjc4n r-1joea0r']")
            head, sep, tail = seguidores.text.partition(' seguidores')
            if head:
                film_info.append(head)
            if not head:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown") 
            
    else:
        film_info.append("unknown") 
        film_info.append("unknown") 
        film_info.append("unknown") 

    if actor2_name != "unknown":

        actor2_link = actor2.get_attribute("href")

        actors_browser.get(actor2_link)
        try:
            facebook_link = actors_browser.find_element_by_link_text('Facebook') #ACTOR2 FACEBOOK
            facebook_link.click()
            likes = actors_browser.find_element_by_xpath("//div[contains(text(), 'personas les gusta esto')]")
            head, sep, tail = likes.text.partition(' personas')
            head2, sep, tail = head.partition('A ')
            if tail:
                film_info.append(tail)
            if not tail:
                film_info.append("unknown")            
        except NoSuchElementException:
            film_info.append("unknown") 

        actors_browser.get(actor2_link)      
        try:
            instagram_link = actors_browser.find_element_by_link_text('Instagram') #ACTOR2 INSTAGRAM
            instagram_link.click()
            seguidores = actors_browser.find_elements_by_xpath("//span[@class='g47SY ']")
            if seguidores:
                film_info.append(seguidores[1].text)
            if not seguidores:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown")

        actors_browser.get(actor2_link)       
        try:
            twitter_link = actors_browser.find_element_by_link_text('Twitter') #ACTOR2 TWITTER
            twitter_link.click()
            time.sleep(1)
            seguidores = actors_browser.find_element_by_xpath("//div[@class='css-1dbjc4n r-1joea0r']")
            head, sep, tail = seguidores.text.partition(' seguidores')
            if head:
                film_info.append(head)
            if not head:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown")         
    
    else:
        film_info.append("unknown") 
        film_info.append("unknown") 
        film_info.append("unknown") 

    if actor3_name != "unknown":

        actor3_link = actor3.get_attribute("href")

        actors_browser.get(actor3_link)
        try:
            facebook_link = actors_browser.find_element_by_link_text('Facebook') #ACTOR3 FACEBOOK
            facebook_link.click()
            likes = actors_browser.find_element_by_xpath("//div[contains(text(), 'personas les gusta esto')]")
            head, sep, tail = likes.text.partition(' personas')
            head2, sep, tail = head.partition('A ')
            if tail:
                film_info.append(tail)
            if not tail:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown") 

        actors_browser.get(actor3_link)    
        try:
            instagram_link = actors_browser.find_element_by_link_text('Instagram') #ACTOR3 INSTAGRAM
            instagram_link.click()
            seguidores = actors_browser.find_elements_by_xpath("//span[@class='g47SY ']")
            if seguidores:
                film_info.append(seguidores[1].text)
            if not seguidores:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown") 

        actors_browser.get(actor3_link)
        try:
            twitter_link = actors_browser.find_element_by_link_text('Twitter') #ACTOR2 TWITTER
            twitter_link.click()
            time.sleep(1)
            seguidores = actors_browser.find_element_by_xpath("//div[@class='css-1dbjc4n r-1joea0r']")
            head, sep, tail = seguidores.text.partition(' seguidores')
            if head:
                film_info.append(head)
            if not head:
                film_info.append("unknown")
        except NoSuchElementException:
            film_info.append("unknown")       
    else:
        film_info.append("unknown") 
        film_info.append("unknown") 
        film_info.append("unknown") 
  
    csv_row.append(film_info)

    return csv_row
```

```python
def header():
    csv_row=[]
    head = ["CODE","ACTOR1_NAME","ACTOR2_NAME","ACTOR3_NAME","ACTOR1_FACEBOOK_LIKES","ACTOR1_INSTAGRAM_LIKES","ACTOR1_TWITTER_LIKES","ACTOR2_FACEBOOK_LIKES","ACTOR2_INSTAGRAM_LIKES","ACTOR2_TWITTER_LIKES","ACTOR3_FACEBOOK_LIKES","ACTOR3_INSTAGRAM_LIKES","ACTOR3_TWITTER_LIKES"]
    csv_row.append(head)
    return csv_row
```

```python
id_films = film_list
id_films = id_films[89926:]  #PARA NO EMPEZAR DESDE EL PRINCIPIO
browser = webdriver.Firefox()

start = time.time()
row = 0
for film in id_films:
    try:
        url = 'https://www.imdb.com/title/'+film+'/?ref_=fn_al_tt_1' 
        browser.get(url)

        with open('film_actors_9.csv', 'a', encoding="utf-8") as f:                

            writer = csv.writer(f,delimiter='\t')
            if row == 0 :
                writer.writerows(header())
            writer.writerows(film_info(film))

        row += 1
        if row % 500 == 0:
            end = time.time()
            print(end-start)
            print(row)

        if row % 200 == 0:
            browser.close()
            time.sleep(1)
            browser = webdriver.Firefox()
            time.sleep(1)
            time.sleep(1)
    except (NoSuchWindowException,WebDriverException) as e:
        
        time.sleep(3)
        browser = webdriver.Firefox()
        browser.get(url)

        with open('film_actors_9.csv', 'a', encoding="utf-8") as f:                

            writer = csv.writer(f,delimiter='\t')
            if row == 0 :
                writer.writerows(header())
            writer.writerows(film_info(film))

        row += 1
        if row % 500 == 0:
            end = time.time()
            print(end-start)
            print(row)

        if row % 200 == 0:
            browser.close()
            time.sleep(1)
            browser = webdriver.Firefox()
            time.sleep(1)
            time.sleep(1)   
```
