distribucion del rating de las películas que no tienen el abstract

```
language = "en"
max_ngram_size = 1
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 1
```

```
descripciones = data.DESCRIPTION.values
```

```
keywords_total = []
for i in tqdm(descripciones):
  
  custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
  keywords = custom_kw_extractor.extract_keywords(i)
  keywords_total = keywords_total + [[a_tuple[0] for a_tuple in keywords]]
```

```
keywords_total =np.squeeze(keywords_total)
```

```
data['Keywords_Description'] = keywords_total
```

```
data.to_csv("imdb_data.csv",sep=",")
```
