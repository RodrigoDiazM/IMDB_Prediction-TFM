```python
import csv   
import pandas as pd
import numpy as np
```

```python
data1 =  pd.read_csv('../data/IMBD_data.csv',  sep='\t', engine='python', header= 0)
data2 =  pd.read_csv('../data/IMBD_data_2.csv',  sep='\t', engine='python', header= 0)
data3 =  pd.read_csv('../data/IMBD_data_3.csv',  sep='\t', engine='python', header= 0)
data4 =  pd.read_csv('../data/IMBD_data_4.csv',  sep='\t', engine='python', header= 0)
data5 =  pd.read_csv('../data/IMBD_data_5.csv',  sep='\t', engine='python', header= 0)
data6 =  pd.read_csv('../data/IMBD_data_6.csv',  sep='\t', engine='python', header= 0)
data7 =  pd.read_csv('../data/IMBD_data_7.csv',  sep='\t', engine='python', header= 0)
data8 =  pd.read_csv('../data/IMBD_data_8.csv',  sep='\t', engine='python', header= 0)
```
